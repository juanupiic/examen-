```python
from collections import defaultdict, deque
import matplotlib.pyplot as plt
import networkx as nx
```

```python
class Grafo:
    def __init__(self):
        self.grafo = defaultdict(list)

    def agregar_arista(self, u, v):
        self.grafo[u].append(v)

    def bfs(self, inicio):
        visitado = set()
        cola = deque()
        recorrido_bfs = []  # Para almacenar el recorrido BFS

        cola.append(inicio)
        visitado.add(inicio)

        while cola:
            nodo_actual = cola.popleft()
            recorrido_bfs.append(nodo_actual)

            for vecino in self.grafo[nodo_actual]:
                if vecino not in visitado:
                    cola.append(vecino)
                    visitado.add(vecino)

        return recorrido_bfs

```

```python
# Ejemplo de uso
grafo = Grafo()
grafo.agregar_arista(0, 1)
grafo.agregar_arista(0, 0)
grafo.agregar_arista(1, 2)
grafo.agregar_arista(2, 0)
grafo.agregar_arista(2, 3)
grafo.agregar_arista(3, 3)

print("Recorrido BFS a partir del nodo 2:")
recorrido = grafo.bfs(2)
print(recorrido)
```

```python


# ... Tu código anterior ...

# Crear un objeto de grafo dirigido para visualizar el grafo
G = nx.DiGraph()

# Agregar nodos al grafo
for nodo in grafo.grafo:
    G.add_node(nodo)

# Agregar aristas al grafo
for nodo, vecinos in grafo.grafo.items():
    for vecino in vecinos:
        G.add_edge(nodo, vecino)

# Dibujar el grafo
pos = nx.spring_layout(G)  # Posiciones de los nodos
nx.draw(G, pos, with_labels=True, node_size=450, node_color='lightblue', font_size=10)

# Ejecutar BFS desde el nodo 2 y resaltar el recorrido
bfs_result = list(grafo.bfs(2))
bfs_edges = [(bfs_result[i], bfs_result[i+1]) for i in range(len(bfs_result)-1)]
nx.draw_networkx_edges(G, pos, edgelist=bfs_edges, edge_color='red', width=5)

# Mostrar el grafo
plt.show()
```
